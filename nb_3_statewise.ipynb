```python
df.drop(["Last_Updated"] , axis = 1 ,inplace = True)
```

```python
df['State_Name'] = df['State_Name'].str.lower()
df
```

```python
shp_gdf = gpd.read_file(r"C:\Users\Deepti Anand\Downloads\india map file")
```

```python
shp_gdf['st_nm'] = shp_gdf['st_nm'].str.lower()
shp_gdf
```

```python
merged = shp_gdf.set_index('st_nm').join(df.set_index('State_Name'))
merged.head()
merged
```

```python
fig , ax = plt.subplots(1 , figsize = (9,9))
ax.axis('off')
ax.set_title ('Statewise Registration count of Unorganized workers' ,
              fontdict = {'fontsize' : '15' , 'fontweight': '3'})

fig = merged.plot( column = 'Registration_count' , 
                  cmap = 'Reds' ,
                  linewidth = 1,
                  ax = ax ,
                  edgecolor = '0.5' ,
                  legend = True)
```

### It is evident that UP accounts for highest number of unorganized workers . Followed by Bihar and Bengal .
